<h1>Data Science</h1>

A question in mind is if the there is a significane in the age range of overdosing between males and females. Mainly if the age range for overdosing in Connecticut from 2012-2022 for females is higher than for males. The plan in mind is to sample 100 males and females from the dataset and then use those data to find the mean and standard error to find the z-value. In the end, I would use a two-sample z-test to find out whether I should reject or not reject the null hypothesis.



<h3>Female</h3>

```python
#sample of 100 people: female
df_fs = df_f.sample(n=100)
df_fs
```

```python
#mean of female sample
fmean = df_fs["Age"].mean()
fmean
```

```python
#standard error of female sample
fse = df_fs["Age"].std()
fse = fse/10
fse
```

<h3>Male</h3>

```python
#sample of 100 people: male
df_ms = df_m.sample(n=100)
df_ms
```

```python
#mean of male sample
mmean = df_ms["Age"].mean()
mmean
```

```python
#standard error of male sample
mse = df_ms["Age"].std()
mse = mse/10
mse 
```

<h3>two sample z-test</h3>

```python
#Using Z-test: 
#Requirements:
    #Population sample 100 > 30
    #Histogram is approx. a normal curve for both male and female
    #The population standard deviation for female is known
    #The population standard deviation for male is known
#H0: The age of overdose in female and male in Connecticut are equal from 2012 to 2022
#HA: The age of overdose in females is significantly higher than in males in Connecticut from 2012 to 2022
#Standard error of Female sample
#Standard error of Male sample
#Mean of Female sample
#Mean of Male sample
#Z-value: (obs difference - ex difference)/ SE difference, -1.255143467617291
#P-value: 0.05
```

```python
#z-value
import math
SEdiff = ((mse * mse) + (fse * fse))
z = (fmean-mmean - 0) / (math.sqrt(SEdiff))
z
```

```python
#finding normcdf
from scipy.stats import norm
nc = 1 - norm.cdf(z)
nc
```

Due to the fact that the z value is greater than 0.05, we do not reject the null hypothesis and there is not a significant difference in age range between the death by overdose of female and male from Conneticut from 2012-2022.
